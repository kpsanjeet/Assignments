Answer :

```python
def find_the_difference(s,t):
    for i in s:
        t = t.replace(i,"",1)
    return t  
        
print(find_the_difference("abcd", "abcde"))
print(find_the_difference("", "y"))
print(find_the_difference("ae", "aea"))
```

Question 2:

Answer :

```python
def count_datatypes(*args):
    d = {int:0, str:0, bool:0, list:0, tuple:0, dict:0}
    for i in args:
        d[type(i)] += 1
    return list(d.values())

print(count_datatypes(1, 45, "Hi", False))
print(count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1))
print(count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23))
print(count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6]))
```

Question 3:

Answer :

```python
def fib_str(n,lst):
    f1,f2 = lst
    fib_s = f1+", "+f2
    for i in range(n-2):
        temp = f2+f1
        fib_s += ", "+temp
        f1 = f2
        f2 = temp
    return fib_s

print(fib_str(3, ["j", "h"]))
print(fib_str(5, ["e", "a"]))
print(fib_str(6, ["n", "k"]))
```

Answer :

```python
def ones_threes_nines(num):
    
    div9, rem9 = divmod(num,9)
    div3, rem3 = divmod(rem9,3)
    div1, rem1 = divmod(rem3,1)
    
    return f"nines:{div9}, threes:{div3}, ones:{div1}"

print(ones_threes_nines(10))
print(ones_threes_nines(15))
print(ones_threes_nines(22))
```

Answer :

```python
def fib(n):
    f = [0,1]
    for i in range(n-1):
        temp = f[~0]+f[~1]
        f.append(temp)
    return f[n]

print(fib(6))
print(fib(1))
print(fib(2))
print(fib(20))
```
